```python
#Name: Shreetu Shrestha
#ID: 2417729

def convert_units(conversion_type, value):
    """
    Converts a value between different units of measurement.

    Parameters:
    conversion_type (str): The type of conversion to perform.
                           Options:
                           '1' - Meters to Feet
                           '2' - Feet to Meters
                           '3' - Kilograms to Pounds
                           '4' - Pounds to Kilograms
                           '5' - Liters to Gallons
                           '6' - Gallons to Liters
    value (float): The numeric value to convert.

    Returns:
    float: The converted value.
    """

    if conversion_type == '1':
        return value * 3.28084
    elif conversion_type == '2':
        return value / 3.28084
    elif conversion_type == '3':
        return value * 2.20462
    elif conversion_type == '4':
        return value / 2.20462
    elif conversion_type == '5':
        return value * 0.264172
    elif conversion_type == '6':
        return value / 0.264172
    else:
        return None


def main():
    print("Unit Conversion Program")
    print("Choose conversion type:")
    print("Length:")
    print("  1. Meters to Feet")
    print("  2. Feet to Meters")
    print("Weight:")
    print("  3. Kilograms to Pounds")
    print("  4. Pounds to Kilograms")
    print("Volume:")
    print("  5. Liters to Gallons")
    print("  6. Gallons to Liters")

    conversion_type = input("Enter your choice (1-6): ")

    try:
        value = float(input("Enter the value to convert: "))
        result = convert_units(conversion_type, value)

        if result is None:
            print("Error: Unsupported conversion type selected.")
        else:
            print(f"Converted value: {result:.4f}")

    except ValueError:
        print("Error: Please enter a valid numeric value.")

main();
```

```python
def calculate_sum(numbers):
    """
    Calculates the sum of a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The sum of the numbers.
    """
    return sum(numbers)


def calculate_average(numbers):
    """
    Calculates the average of a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The average of the numbers.
    """
    return sum(numbers) / len(numbers)


def find_maximum(numbers):
    """
    Finds the maximum value in a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The maximum value.
    """
    return max(numbers)


def find_minimum(numbers):
    """
    Finds the minimum value in a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The minimum value.
    """
    return min(numbers)


def main():
    print("Choose an operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    choice = input("Enter your choice (1-4): ")

    try:
        user_input = input("Enter numbers separated by spaces: ")
        numbers = list(map(float, user_input.split()))

        if len(numbers) == 0:
            raise ValueError("Empty list is not allowed.")

        if choice == '1':
            result = calculate_sum(numbers)
        elif choice == '2':
            result = calculate_average(numbers)
        elif choice == '3':
            result = find_maximum(numbers)
        elif choice == '4':
            result = find_minimum(numbers)
        else:
            print("Invalid operation selected.")
            return

        print("Result:", result)

    except ValueError:
        print("Error: Please enter valid numeric values and avoid empty input.")


if __name__ == "__main__":
    main()
```

```python
def extract_every_other(lst):
  """
  Extracts every other element from a list.
  Parameters:
  lst: Input list
  Returns:
  list: A new list containing every other element
  """
  return lst[::2]


def get_sublist(lst, start, end):
  """
  Returns a sublist from start index to end index (inclusive)

  Parameters:
  lst: Input list
  start (int): Starting index
  end (int): Ending index

  Returns:
  list: The extracted sublist
  """
  return lst[start:end+1]


def reverse_list(lst):
  """
  Reverses a list using slicing

  Parameters:
  lst (list): Input list

  Returns:
  list: Reversed list
  """
  return lst[::-1]


def remove_first_last(lst):
  """
  Removes the first and last elements from a list.

  Parameters:
  lst (list): Input list.

  Returns:
  list: List without first and last elements.
  """
  return lst[1:-1]


def get_first_n(lst, n):
    """
    Returns the first n elements of a list.

    Parameters:
    lst (list): Input list.
    n (int): Number of elements to extract.

    Returns:
    list: First n elements.
    """
    return lst[:n]


def get_last_n(lst, n):
    """
    Returns the last n elements of a list.

    Parameters:
    lst (list): Input list.
    n (int): Number of elements to extract.

    Returns:
    list: Last n elements.
    """
    return lst[-n:]


def reverse_skip(lst):
    """
    Returns every second element starting from the second-to-last element,
    moving backward.

    Parameters:
    lst (list): Input list.

    Returns:
    list: Modified list as described.
    """
    return lst[-2::-2]


print(get_last_n([1, 2, 3, 4, 5, 6], 2))
```

```python
def flatten(lst):
  """
  Returns a flattened list

  Parameters:
  lst (list): Input list

  Returns:
  list: Flattened list
  """
  flat_list = []
  for sublist in lst:
      for item in sublist:
          flat_list.append(item)
  return flat_list

def access_nested_element(lst, indices):
  """
  Accesses an element in nested list using a list of indices

  Parameters:
  lst (list): Nested list
  indices (list): List of indices

  Returns:
  element at the specidied position
  """

  element = lst
  for index in indices:
    element = element[index]
  return element

def sum_nested(lst):
  """
  Cakcukated the sum of all elements in a nested list

  Parameters:
  lst (list): Nested list

  Returns:
  int/float: total sum of all elements
  """

  total = 0
  for item in lst:
    if isinstance(item,list):
      total += sum_nested(item)
    else:
      total += item
  return total

def remove_element(lst, elem):
  """
  Removes all occurances of element from a nested list

  Parameters:
  lst (list): Nested list
  elem: element ot remove

  Returns:
  list: modified list without element
  """

  new_list = []
  for item in lst:
    if isinstance(item, list):
      new_list.append(remove_element(item, elem))
    elif item != elem:
      new_list.append(item)
  return new_list

def find_max(lst):
  """
  Finds maximum element in a nested list

  Parameters:
  lst(list): Nested list

  Returns:
  int/float: Maximun element
  """
  max_value= None

  for item in lst:
    if isinstance(item, list):
      value = find_max(item)
    else:
      value = item

    if max_value is None or value > max_value:
      max_value = value
  return max_value

def count_occurrences(lst, elem):
  """
  counts occurences of elements in list

  Parameters:
  lst (list): Input list
  elem: element to count

  Returns:
  int: number of occurences
  """
  count = 0
  for item in lst:
    if isinstance(item,list):
      count += count_occurrences(item, elem)
    elif item == elem:
      count += 1
  return count

def deep_flatten(lst):
  """
  Flattens a deeply nested list into a single list.

  Parameters:
  lst (list): Deeply nested list.

  Returns:
  list: Fully flattened list.
  """
  flat_list = []
  for item in lst:
      if isinstance(item,list):
          flat_list.extend(deep_flatten(item))
      else:
          flat_list.append(item)
  return flat_list

def average_nested(lst):
  """
  calculated the average of all elements in a nested list
  parameters:
  lst(list): nested list

  Returns:
  float: Average value
  """
  flat = deep_flatten(lst)
  return sum(flat)/ len(flat)

lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

print(flatten([[1, 2], [3, 4], [5]]))
print(access_nested_element(lst, [1, 2]))
print(sum_nested([[1, 2], [3, [4, 5]], 6]))
print(remove_element([[1, 2], [3, 2], [4, 5]], 2))
print(find_max([[1, 2], [3, [4, 5]], 6]))
print(count_occurrences([[1, 2], [2, 3], [2, 4]], 2))
print(deep_flatten([[[1, 2], [3, 4]], [[5, 6], [7, 8]]]))
print(average_nested([[1, 2], [3, 4], [5, 6]]))
```
